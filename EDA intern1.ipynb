# Exploratory Data Analysis on the Titanic Dataset

The Titanic dataset is one of the most widely used datasets in data science and machine learning. It provides a detailed account of passengers aboard the RMS Titanic, a British passenger liner that tragically sank on April 15, 1912. The dataset includes passenger attributes such as age, gender, class, fare, and survival status, offering an opportunity to explore demographic and social factors influencing survival rates.

# This analysis focuses on:

Data Cleaning: Addressing missing values, data type conversions, and preparing the dataset.
Exploratory Data Analysis (EDA): Visualizing and identifying patterns, trends, and relationships within the data.
Insights: Gaining a deeper understanding of the factors influencing survival, such as gender, age, and socioeconomic status.
Through this analysis, we aim to uncover meaningful insights and enhance our understanding of the tragic event using a data-driven approach. The findings also provide a foundational understanding of EDA techniques for beginner data analysts.

# How does gender influence the survival rate of passengers on the Titanic

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x=df['Sex'],y=df['Survived'],palette='Set2')
plt.show()
```

# The bar plot shown compares the survival rates between male and female passengers on the Titanic. Here's what the visualization indicates:

Observation:

Female passengers have a significantly higher survival rate compared to male passengers.
The average survival rate for males is much lower (around 20â€“30%), while females exhibit a survival rate of approximately 70â€“80%.
Reason:

The survival disparity can be attributed to the "Women and children first" evacuation policy that was followed during the Titanic disaster, prioritizing females for lifeboats.
Insights:

Gender played a crucial role in determining survival.
This suggests that gender could be a strong feature for predicting survival in machine learning models built on this dataset.
Further Exploration:

Add another variable, such as Pclass or Age, to examine if survival rates differ by class or age group within each gender.


# What does the age distribution of Titanic passengers reveal

```python
sns.histplot(df['Age'], bins=20, kde=True, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# Observation:

The histogram shows the frequency distribution of passenger ages.
Most passengers are in their 20s and 30s, with the peak frequency (mode) around 30 years.
There is a noticeable decline in the frequency of older passengers (above 50 years).
KDE (Kernel Density Estimate):

The blue KDE curve provides a smooth estimate of the age distribution.
It highlights the concentration of younger passengers while tapering off for older age groups.
Insights:

The majority of Titanic passengers were young adults (20â€“40 years).
There is also a smaller population of children and teenagers, as evident from the tail extending towards lower ages (below 20).

# Survival by Passenger Class

```python
sns.countplot(x='Pclass', hue='Survived', data=df, palette='pastel')
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(['Not Survived', 'Survived'])
plt.show()
```

# Observations:
The bar plot depicts the number of survivors and non-survivors across different passenger classes.
It uses blue and orange to represent "Not Survived" and "Survived," respectively.
The x-axis shows the passenger classes (1, 2, and 3), while the y-axis represents the count.

# How does survival status correlate with fare and age

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, palette='viridis')
plt.title('Age vs Fare (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

# Observations:
The scatter plot visualizes the relationship between passengers' ages and fares, colored by survival status.
There seems to be a concentration of data points for lower fares, regardless of age.
Younger passengers (under 20) and those in the middle age group (20-50) dominate the dataset.
The survival rate (Survived=1, green points) seems to vary slightly across different fare values and ages, with some clustering at higher fare values.
Outliers are visible, such as individuals with very high fares (over 500), and they seem to have survived.

# Does a higher fare consistently lead to better survival odds across all passenger classes?

```python
numerical_columns = ['Age', 'Fare', 'Pclass', 'Survived']
sns.pairplot(df[numerical_columns], hue='Survived', palette='husl', diag_kind='kde')
plt.show()
```

# Observations:
The plot is a pair plot, which provides pairwise scatter plots and histograms for multiple variables, colored by the "Survived" column.
Diagonal plots represent the distribution (histograms and KDEs) for each variable, separated by survival status:
Age shows a peak for younger individuals, with a slight difference between survivors and non-survivors.
Fare distribution is heavily skewed, with most passengers paying lower fares.
Passenger Class (likely represented numerically) has clear clustering in distinct groups.
Off-diagonal plots display scatter plots between variable pairs:
There is no clear linear relationship between Age and Fare, but survival status is noticeable in specific clusters (e.g., younger passengers with higher fares show more survival).
Passenger Class exhibits distinct clustering with survival patterns.
Non-survivors (Survived=0, pink points) dominate most variable distributions compared to survivors (Survived=1, green points).

# Conclusion:
The analysis confirmed that survival on the Titanic was influenced significantly by socio-economic factors such as fare and passenger class. Younger passengers and those in higher classes had better chances of survival.
